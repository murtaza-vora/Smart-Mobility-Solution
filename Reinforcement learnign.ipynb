```python
import gym
from gym import spaces
import numpy as np

class TrafficManagementEnv(gym.Env):
    def __init__(self):
        super(TrafficManagementEnv, self).__init__()

        # Define action and observation spaces
        self.action_space = spaces.Discrete(num_actions)  # Define the number of possible actions (e.g., control traffic lights)
        self.observation_space = spaces.Box(low=0, high=1, shape=(num_observations,), dtype=np.float32)  # Define the observation space (e.g., traffic flow rates)

        # Initialize state
        self.state = self.reset()

    def step(self, action):
#         # Implement the environment dynamics
#         # Update the state based on the chosen action
#         # Compute the reward based on the updated state
#         # Return the next state, reward, and whether the episode is done

    def reset(self):
#         # Reset the environment to its initial state
#         # Return the initial state

# Define hyperparameters
num_actions = 4  # Number of possible actions (e.g., control traffic lights)
num_observations = 6  # Number of observations (e.g., traffic flow rates)
num_episodes = 1000  # Number of training episodes
max_steps_per_episode = 1000  # Maximum number of steps per episode

# Initialize the environment
env = TrafficManagementEnv()

# Initialize the Q-table
Q = np.zeros([num_observations, num_actions])

# Define hyperparameters for the Q-learning algorithm
learning_rate = 0.1
discount_factor = 0.99
exploration_rate = 1.0
max_exploration_rate = 1.0
min_exploration_rate = 0.01
exploration_decay_rate = 0.01

# Q-learning algorithm
for episode in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0

    for step in range(max_steps_per_episode):
        # Choose an action (exploration-exploitation trade-off)
        if np.random.uniform(0, 1) < exploration_rate:
            action = env.action_space.sample()  # Exploration
        else:
            action = np.argmax(Q[state, :])  # Exploitation

        # Perform the chosen action and observe the next state and reward
        next_state, reward, done, _ = env.step(action)

        # Update the Q-table
        Q[state, action] = Q[state, action] + learning_rate * (reward + discount_factor * np.max(Q[next_state, :]) - Q[state, action])

        state = next_state
        total_reward += reward

        if done:
            break

    # Update exploration rate
    exploration_rate = min_exploration_rate + (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate * episode)

    # Print the total reward achieved in this episode
    print("Episode:", episode, "Total Reward:", total_reward)

# Optionally, you can test the trained agent here
```

```python
import numpy as np

# Define the environment (traffic intersection)
class TrafficIntersection:
    def __init__(self):
        self.num_actions = 2  # Two possible actions: green or red light
        self.states = [0, 1]  # Two possible states: north-south traffic or east-west traffic
        self.state = np.random.choice(self.states)  # Initial state
        self.q_table = np.zeros((len(self.states), self.num_actions))  # Q-table initialization
        self.alpha = 0.1  # Learning rate
        self.gamma = 0.9  # Discount factor
        self.epsilon = 0.1  # Epsilon-greedy parameter

    def choose_action(self):
        if np.random.uniform(0, 1) < self.epsilon:
            # Exploration: choose a random action
            return np.random.choice(self.num_actions)
        else:
            # Exploitation: choose the action with the highest Q-value
            return np.argmax(self.q_table[self.state])

    def update_q_table(self, action, reward, next_state):
        # Q-learning update rule
        old_q_value = self.q_table[self.state, action]
        td_target = reward + self.gamma * np.max(self.q_table[next_state])
        self.q_table[self.state, action] += self.alpha * (td_target - old_q_value)

    def step(self, action):
        # Simulate traffic flow and return the reward
        if self.state == 0:  # North-South traffic
            if action == 0:  # Green light for North-South traffic
                reward = 1  # Positive reward for smooth traffic flow
            else:  # Red light for North-South traffic
                reward = -1  # Negative reward for congestion
        else:  # East-West traffic
            if action == 1:  # Green light for East-West traffic
                reward = 1  # Positive reward for smooth traffic flow
            else:  # Red light for East-West traffic
                reward = -1  # Negative reward for congestion
        return reward

    def reset(self):
        self.state = np.random.choice(self.states)


```

```python
# Initialize the traffic intersection environment
intersection = TrafficIntersection()

# Train the Q-learning agent
num_episodes = 1
for episode in range(num_episodes):
    intersection.reset()
    done = False
    while not done:
        action = intersection.choose_action()
        reward = intersection.step(action)
        next_state = np.random.choice(intersection.states)
        intersection.update_q_table(action, reward, next_state)
        intersection.state = next_state

```

```python
import time

# Initialize the traffic intersection environment
intersection = TrafficIntersection()

# Train the Q-learning agent
num_episodes = 1
start_time = time.time()  # Start timing

for episode in range(num_episodes):
    intersection.reset()
    done = False
    while not done:
        action = intersection.choose_action()
        reward = intersection.step(action)
        next_state = np.random.choice(intersection.states)
        intersection.update_q_table(action, reward, next_state)
        intersection.state = next_state

end_time = time.time()  # End timing
processing_time = end_time - start_time
print("Processing time:", processing_time, "seconds")
```

```python
# Evaluate the learned policy
total_rewards = 0
num_eval_episodes = 100
for _ in range(num_eval_episodes):
    intersection.reset()
    done = False
    while not done:
        action = np.argmax(intersection.q_table[intersection.state])
        reward = intersection.step(action)
        total_rewards += reward
        next_state = np.random.choice(intersection.states)
        intersection.state = next_state

average_reward = total_rewards / num_eval_episodes
print("Average reward:", average_reward)
```
